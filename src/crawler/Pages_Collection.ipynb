```python
path = "/Users/kittnguyen/Library/Mobile Documents/com~apple~CloudDocs/Phân tích và thiết kế thực nghiệm/Đồ án cuối kỳ/URL_Dataset.csv"
df = pd.read_csv (path).drop ('Unnamed: 0', axis=1)
df
```

```python
# Chuyển DataFrame thành list of dict

class PageSchema (BaseModel):
    URL : str
    Image: str
    Title: str
    Details: str
    Search: str
    Region: str

for idx, row in df.iterrows ():
    data = {
        'URL': row ['link'],
        'Image': row ['image_url'],
        'Title': row ['title'],
        'Details': row ['details'],
        'Search': row ['Search_Info'],
        'Region': row ['Region'],
    }
    try:
        page = PageSchema (**data)
        Pages.insert_one (page.dict ())
        print (f"Insert dòng {idx} thành công")
    except ValidationError as e:
        print (f"Lỗi dòng {idx}:", e.json ())


# records = df.to_dict (orient='records')
# for idx, record in enumerate (tqdm (records, desc='Inserting records')):
#     try:
#         page = PageSchema (**record) # validate
#         Pages.insert_one (page.dict ())
#     except ValidationError as e:
#         print (f"Validation error at record {idx}: {e}")
```
